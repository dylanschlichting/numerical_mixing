```python
mnum_diff = (mnum_ss_30-mnum_ss_10[3:-3][::3]).values #convert from xarray
tend_diff = ((ds2dt_30-ds2dt_10[3:-3][::3]))
diffusion_diff = (s2_diff_flux_30-s2_diff_flux_10[3:-3][::3]).values 
adv_diff = (s2_advection_30-s2_advection_10[3:-3][::3]).values 
surf_diff = (s2_surface_30-s2_surface_10[3:-3][::3]).values
chi_diff = (chi_online_30-chi_online_10[3:-3][::3]).values


# Construct pandas dataframe to for easy integration into existing packages and plotting
data = {'delta_tend': tend_diff,
        'delta_adv': adv_diff,
        'delta_surf': surf_diff,
        'delta_hdiff': diffusion_diff,
        'delta_chi': chi_diff,
        'delta_mnum': mnum_diff
        }
df = pd.DataFrame(data)
```
